```python
#total revenue generated by male and female customers
q1 = '''
select gender,
sum("purchase_amount_(usd)") as revenue
from customer
group by gender
'''

pd.read_sql(q1, engine)
```

```python
#customers with discount whose purchase is closest to average purchase
q2 = '''
select 
    customer_id,
    "purchase_amount_(usd)"
from customer
where discount_applied='Yes'
order by abs(
    "purchase_amount_(usd)" - 
    (select avg("purchase_amount_(usd)") from customer)
)
'''

pd.read_sql(q2, engine)
```

```python
#top 5 products with highest average review rating
q3 = '''
select 
    item_purchased,
    round(avg(review_rating)::numeric, 2) as "Average Product Rating"
from customer
group by item_purchased
order by avg(review_rating) desc
limit 5
'''

pd.read_sql(q3, engine)
```

```python
#average purchase amount for each shipping type
q4 = '''
select 
    shipping_type,
    round(avg("purchase_amount_(usd)")::numeric, 2) as average_purchase
from customer
where shipping_type in ('Standard', 'Express')
group by shipping_type
'''

pd.read_sql(q4, engine)
```

```python
#total revenue and average spend for each subscription status
q5 = '''
select 
    subscription_status,
    count(customer_id) as total_customers,
    round(avg("purchase_amount_(usd)")::numeric, 2) as avg_spend,
    round(sum("purchase_amount_(usd)")::numeric, 2) as total_revenue
from customer
group by subscription_status
order by total_revenue, avg_spend desc
'''

pd.read_sql(q5, engine)
```

```python
#top 5 products with highest discount rate
q6 = '''
select 
    item_purchased,
    round(
        sum(
            case 
                when discount_applied='Yes' then 1 
                else 0 
            end
        ) * 100.0 / count(*), 
        2
    ) as discount_rate
from customer
group by item_purchased
order by discount_rate desc
limit 5
'''

pd.read_sql(q6, engine)
```

```python
#segment customers based on their previous purchases into three categories: New, Returning, and Loyal. Then, count the number of customers in each segment.
q7 = '''
with customer_type as (
    select 
        customer_id,
        previous_purchases,
        case
            when previous_purchases = 1 then 'New'
            when previous_purchases between 2 and 10 then 'Returning'
            else 'Loyal'
        end as customer_segment
    from customer
)

select 
    customer_segment,
    count(*) as "Number of Customers"
from customer_type
group by customer_segment
'''

pd.read_sql(q7, engine)
```

```python
#top 3 most purchased items in each category
q8 = '''
with item_counts as (
    select
        category,
        item_purchased,
        count(customer_id) as total_orders,
        row_number() over(
            partition by category
            order by count(customer_id) desc
        ) as item_rank
    from customer
    group by category, item_purchased
)

select
    item_rank,
    category,
    item_purchased,
    total_orders
from item_counts
where item_rank <= 3
'''

pd.read_sql(q8, engine)
```

```python
#are customers with more than 5 previous purchases more likely to be subscribed to a subscription plan?
q9 = '''
select
    subscription_status,
    count(customer_id) as repeat_buyers
from customer
where previous_purchases > 5
group by subscription_status
'''

pd.read_sql(q9, engine)
```

```python
#revenue contribution from different age groups
q10 = '''
select
    age_group,
    sum("purchase_amount_(usd)") as total_revenue
from customer
group by age_group
order by total_revenue desc
'''

pd.read_sql(q10, engine)
```
